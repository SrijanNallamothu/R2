```python
#Functions

def remove_punct(text):
    string_without_punc = ""
    for line in text:
        for char in line:
            string_without_punc += ''.join(' ' if c in string.punctuation else c for c in line)
    return string_without_punc

def remove_nextLine(text):
    string_without_line_breaks = ""
    for line in text:
        stripped_line = line.replace("\n", "")
        string_without_line_breaks += stripped_line
    return string_without_line_breaks

def remove_tab(text):
    string_without_tabs = ""
    for line in text:
        stripped_line = line.replace("\t", "")
        string_without_tabs += stripped_line
    return string_without_tabs

def tokenize(text):
    tokens = text.split()
    return tokens

def remove_stopWords(list):
    text = [word for word in list if word not in stopword]
    return text

from bs4 import BeautifulSoup

def remove_html(text):
    cleantext = BeautifulSoup(text, "html.parser").text
    return cleantext

import re
def remove_urls (text):
    clean_text = re.sub(r'(https|http)?:\/\/(\w|\.|\/|\?|\=|\&|\%)*\b', '',text, flags=re.MULTILINE)
    return(clean_text)

def clean_text(text):
    clean_text = remove_punct(text);
    clean_text = remove_urls(clean_text)
    clean_text = remove_html(clean_text)
    clean_text = remove_nextLine(clean_text)
    clean_text = remove_tab(clean_text)
    clean_text = tokenize(clean_text)
    clean_text = remove_stopWords(clean_text)
    return clean_text
```
